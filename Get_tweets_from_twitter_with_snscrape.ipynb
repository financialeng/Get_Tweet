```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

Now it's time to write our search words like what we do on twiiter exploreScreenshot 2022-12-03 163256.png

I used #BTC 

Then I choosed how many last tweets I want, for example I want 100 last #BTC tweets so ( if i>100 ) break

tweet.date  is date of tweet

tweet.likeCount  counts likes of tweet

tweet.retweetCount  counts retweets of tweet

and so on , you can find replycount , quotecount , sourcelabel , username , ID , language , content or tweet text , URL and media of tweet on the code here :



```python
attributes_container = []

for i,tweet in enumerate(sntwitter.TwitterSearchScraper('#BTC').get_items()):
    if i>100:
        break
    attributes_container.append([tweet.date, tweet.likeCount, tweet.retweetCount, tweet.replyCount, tweet.quoteCount , tweet.sourceLabel, tweet.user.username, tweet.id, tweet.lang, tweet.content, tweet.url, tweet.media])
    
```

Let's create a DataFrame for the tweets list above :

```python
tweets_df = pd.DataFrame(attributes_container, columns=["Date Created", "Likes","Retweets","Replies","Quotes","Source of Tweet","Username","ID","Lang", "Tweets","URL","Media"])
```

Or how to collect tweets of #BTC from 2022/11/01 to 2022/11/10  :
( choose i propely to get all the tweets you want )

```python
attributes_container = []

for i,tweet in enumerate(sntwitter.TwitterSearchScraper('#BTC lang:en until:2022-11-10 since:2022-11-01').get_items()):
    if i>100000:
        break
    attributes_container.append([tweet.date, tweet.likeCount, tweet.retweetCount, tweet.replyCount, tweet.quoteCount , tweet.sourceLabel, tweet.user.username, tweet.id, tweet.lang, tweet.content, tweet.url, tweet.media])
    
```
